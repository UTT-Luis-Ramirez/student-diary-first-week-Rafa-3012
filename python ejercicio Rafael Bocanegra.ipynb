#**Configuracion de ambiente**

#**Importacion de librerias**



```python
import pandas as pd
data = pd.read_csv ("/content/SoftwareProject_DataSet.csv")
```

#**Verificacion del DataSet**


```python
data.head()
```

#**Obtener tamaño poblacion**

```python
population_size = len(data)
print("The population size size is over:", population_size)
```

#**Creacion de muestra aleatoria simple de 50 registros**

```python
sample_size = 50
simple_sample = data.sample(sample_size)
```

#**Revision de creacion de la muestra**

```python
display(simple_sample)
```

#**Obtener el tamaño de la muestra**


```python
simple_size = len (simple_sample)
print("The simple size is over" , simple_size)
```

#**Descripcion de la muestra**

```python
sample_description = simple_sample[["Costo (USD)", "Tiempo de desarrollo (horas)"]]
print("Descripcion de la muestra")
display(sample_description)
```

#**Muestreo Aleatorio sistematico**

#**Seleccion sistematica de una muestra**

#**Definir el intarvalo k(tamaño poblacion / muestra deseado)

```python
sample_size = 80
interval = population_size // sample_size
```

#**Seleccion de muestra sistematica**

```python
systematic_sample = data.iloc[::interval]
```

#**Verificacion de la muestra**

```python
print("Muestra o venida a traves de muestreo aleatorio sistematico")
display(systematic_sample)
```

#**Analisis de subpoblaciones**

---



#**Dividir la poblacion por tipo de proyecto

---



```python
grouped_by_project_type = data.groupby("Tipo de proyecto")
grouped_by_project_type.head()
```

#**Calcular el promedio de lineas de codigo y bugs para cada subpoblacion**

---



```python
group_analysis = grouped_by_project_type[["Líneas de código" , "Número de bugs" , "Costo (USD)" ]].mean()
display(group_analysis)
```

#**Dividir la poblacion por tipo de lenguaje**

---



```python
grouped_by_language = data.groupby("Idioma de programación")
grouped_by_language.head()
```

#**Calcular el promedio de la satisfaccion del clente por lenguaje**

```python
group_analysis = grouped_by_language[["Satisfacción del cliente"]].mean()
display(group_analysis)
```

#**Calcular el promedio de antiguedad del desarrollador por lenguaje**

```python
group_analysis = grouped_by_language[["Antigüedad del desarrollador (años)"]].mean()
display(group_analysis)
```

#**Calcular el promedio del tiempo de desarrollo por lenguaje**

```python
group_analysis = grouped_by_language[["Tiempo de desarrollo (horas)"]].mean()
display(group_analysis)
```
